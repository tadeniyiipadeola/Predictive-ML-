```python
from csv import writer

def append_list_as_row(file_name, list_of_elem):
    # Open file in append mode
    with open(file_name, 'a+', newline='') as write_obj:
        # Create a writer object from csv module
        csv_writer = writer(write_obj)
        # Add contents of list as last row in the csv file
        csv_writer.writerow(list_of_elem)
```

```python
# List of strings
row_contents = [Job_info, shrimp_number, BHA_RPM_mean, BHA_RPM_max, BHA_RPM_min, BHA_RPM_std, RPM_over_200, percent_of_RPM_over_200,  
                temp_mean, temp_max, temp_min, temp_std, 
                Z_Gs_mean, Z_Gs_max, Z_Gs_min, Z_Gs_std, 
                XY_Gs_mean, XY_Gs_max, XY_Gs_min, XY_Gs_std]
# Append a list as new line to an old csv file
append_list_as_row('SHRIMP Report.csv', row_contents)
```
